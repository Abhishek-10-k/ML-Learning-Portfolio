# Exploratory Data Analysis on Cleveland Heart Disease Dataset


## Intoduction

### Overview of the Cleveland Heart Disease Dataset

#### The Cleveland Heart Disease dataset is a collection of medical information about 303 patients. It is used to study and predict whether a person has heart disease (1 = yes, 0 = no). There are 14 pieces of information (or features) about each patient, which include things like:


1.   Age: The person’s age.
2.   Sex: Whether the person is male (1) or female (0).

1.   Chest Pain Type: Different types of chest pain the person has experienced (4 types).
2.   Resting Blood Pressure: Their blood pressure when they are at rest.

1.   Cholesterol: The amount of cholesterol in their blood.
2.   Fasting Blood Sugar: Whether their blood sugar level is above 120 mg/dl after fasting.

1.   Resting ECG Results: Results from a heart test that checks electrical activity.
2.   Max Heart Rate: The highest heart rate they reached during physical activity.

1.   Exercise-Induced Angina: Whether they feel chest pain during exercise.
2.   Oldpeak: A measure of changes in heart rate during exercise.

1.   Slope of ST Segment: The direction of changes in the heart rate after exercise.
2.   Major Vessels: The number of major blood vessels visible in a scan.

1.   Thalassemia: A blood disorder, with different types noted.

Doctors and researchers use this information to understand heart disease and make predictions about a patient’s heart health.















### Objectives of the EDA

#### The main goals of Exploratory Data Analysis (EDA) are:

1.   **Understand the data:** Get to know the data’s structure and contents.
2.   **Find patterns:** Discover relationships and trends.

1.   **Spot errors:** Identify any anomalies or outliers.
2.   **Check assumptions:** Verify any assumptions about the data.

1.   **Guide analysis:** Decide on the next steps for deeper analysis.

## Visualization

```python
# Categorical variables

categorical_columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']

# Numerical variables

numerical_columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
```

### 1. Scatter Plots

```python
fig = px.scatter(df, x='age', y='chol', color='target', title='Scatter Plot of Age vs Cholesterol')
fig.show()
```

```python
fig = px.scatter(df, x='thalach', y='trestbps', color='target', title='Scatter Plot of Thalach vs Trestbps')
fig.show()
```


*   **Purpose of the visual :** A scatter plot is a type of graph used to show the relationship between two variables.

*   **Insights :**

1.   No clear positive or negative trend.
2.   Appears to be nonlinear.

1.   The points are widely scattered, indicating a weak relationship.
2.   There are points that are far from the main cluster, which might be outliers.

1.   The ‘target’ variable adds another layer of information, showing how it varies with the other 2 variables.










### 2. Pair Plot

```python
sns.pairplot(df[numerical_columns])
```


*   **Pair plot :** A pair plot is a type of graph that shows the relationships between multiple variables in a dataset. It creates a grid of plots where each plot shows the relationship between a pair of variables. The diagonal plots show the distribution of each variable.
*   **Insights :**

1.   Age appears to be normally distibuted.
2.   Trestbps & chol shows a slight right skew.

1.   Thalach is Roughly normally distributed.
2.   Oldpeak is Skewed to the right, indicating most values are low with a few high outliers.

1.   As age increases, maximum heart rate tends to decrease, which shows a slight negative correlation.
2.   Older individuals tend to have higher ST depression, which shows slight positive correlation.

1.   Some scatter plots show outliers, particularly in ‘oldpeak’.
2.   Most pairs of variables do not show strong correlations, indicating low multicollinearity.









### 3. Box Plots

```python
for col in numerical_columns:
  fig = px.box(df, x='target', y=col, title=f'Box Plot of {col} by Target')
  fig.show()
```




*   **Purpose of the visual :** A box plot (or box-and-whisker plot) is a graph that shows the distribution of a dataset. It displays the data’s minimum, first quartile (Q1), median, third quartile (Q3), and maximum values. Along with that it also identifies the outliers in the datset.
*   **Insights :**

1.   The presence of outliers indicates a need for careful outlier treatment to avoid overfitting.
2.   Features with significant differences in medians between classes can be good candidates for feature selection, as they indicate a potential relationship with the target variable.





### 4. Histogram

```python
for col in numerical_columns:
  fig = px.histogram(df, x=col, title=f'Histogram of {col}')
  fig.show()
```

*   **Purpose of the Visual :** A histogram is a type of graph that shows the distribution of a dataset. It looks like a bar chart but is used for continuous data.
*   **Insights :**

1.   Age appears to be normally distibuted.
2.   Trestbps & chol shows a slight right skew.

1.   Thalach is Roughly normally distributed.
2.   Oldpeak is Skewed to the right, indicating most values are low with a few high outliers.

1.   The peak of the histograms indicates where most data points are concentrated, giving an idea of the central tendency.
2.   All the histograms are unimodal.

1.   The spread of the bars indicates the variability in the data.













### 5. Correlation heatmap

```python
corr_matrix = df[numerical_columns].corr()
fig = px.imshow(corr_matrix, text_auto=True, title='Correlation Matrix of Numerical Columns')
fig.show()
```


*   **Purpose of the Visual :** A correlation heatmap is a visual tool that shows the relationship between multiple variables using colors. Each cell in the heatmap represents the correlation between two variables.
*   **Insights :**

1.   Age and Thalach: As age increases, the maximum heart rate achieved tends to decrease. This is a moderate negative correlation, suggesting that older individuals generally have lower maximum heart rates.
2.   Age and Trestbps: There is a slight positive correlation, indicating that as age increases, resting blood pressure tends to increase slightly.

1.   In this matrix, none of the correlations are extremely high, so all features might still be useful.










### 6. Violin Plot

```python
for col in numerical_columns:
  fig = px.violin(df, x='target', y=col, title=f'Violin Plot of {col} by Target')
  fig.show()
```


*   **Purpose of the visual :** A violin plot is a type of graph that shows the distribution of numerical data. It’s similar to a box plot but also includes a rotated density plot on each side, which looks like a violin.
*   **Insights :**

1.   It basically shows at what values of y, most of the data points are concentrated.
2.   Example : In age vs target (1) violin plot, most of the data is concentrated around the median.









### 7. Count Plot

```python
for col in categorical_columns:
  fig = px.histogram(df, x=col, title=f'Frequency Distribution of {col}')
  fig.show()
```



*   **Purpose of the visual :** A count plot is a type of bar graph used to show the number of occurrences for each category in a dataset.
*   **Insights :**

1.   Male are in much larger numbers in comparison to females.
2.   Class label 0 & 1 have approximately similar representation in the dataset which shows data is unbiased.







### 8. Bar Plot

```python
for cat_col in categorical_columns:
  for num_col in numerical_columns:
    mean_df = df.groupby(cat_col)[num_col].mean().reset_index()
    fig = px.bar(mean_df, x=cat_col, y=num_col, title=f'Mean of {num_col} by {cat_col}')
    fig.show()
```


*   **Purpose of the visual :** A bar plot (or bar graph) is a type of chart that represents data with rectangular bars. Each bar’s length or height is proportional to the value it represents.
*   **Insights :**

1.   Spot Differences: Identify if certain features show significant differences across categories.

1.   In this dataset 'oldpeak' shows significance differnce across various categorical variables.








### 9. Density Plot

```python
for col in numerical_columns:
  fig = px.histogram(df, x=col, nbins=50, histnorm='probability density', title=f'Probability Density Function of {col}')
  fig.show()
```


*   **Purpose of the visual :** A probability density plot is a graph that shows the likelihood of different outcomes for a continuous random variable.
*   **Insights :**



1.   Similar to histogram it help in identifying the data distribuution, and all the insights of histogram are applicable to pdf curve.





### 10. Pairwise Plot with Hue

```python
pair_plot = sns.pairplot(df, vars=numerical_columns, hue='target', palette='viridis')
pair_plot.fig.suptitle('Pair Plot of Numerical Columns Colored by Target', y=1.02)
plt.show()
```


*   **Purpose of the visual :** Creating a pair plot with a hue parameter in Seaborn allows you to color the plot aspects based on the values of a specific variable.
*   **Insights :**



1.   The hue in the pair plot adds a valuable dimension by showing how the target variable interacts with the other features. It helps in identifying patterns, correlations, and potential predictors that might not be obvious from the variables alone.
